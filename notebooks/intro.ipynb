# Topics per Class Using BERTopic

20 Newsgroups データセットを使い、クラスごとのトピック分布を分析する。

参考: [Topics per Class Using BERTopic](https://towardsdatascience.com/topics-per-class-using-bertopic-252314f2640/)

```python
from bertopic import BERTopic
from sklearn.datasets import fetch_20newsgroups

data = fetch_20newsgroups(subset='all', remove=('headers', 'footers', 'quotes'))
docs = data["data"]
classes = [data["target_names"][i] for i in data["target"]]
```

```python
topic_model = BERTopic(verbose=True)
topics, probs = topic_model.fit_transform(docs)
```

```python
topics_per_class = topic_model.topics_per_class(docs, classes=classes)
topics_per_class.head(10)
```

```python
topic_model.visualize_topics_per_class(topics_per_class, top_n_topics=10)
```
